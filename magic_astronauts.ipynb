```python
import csv

path = 'astronauts.csv'
file = open(path)

data = csv.reader(file)

class Astronauts:
    """ Astronauts Class """
    
    def __init__(self, name, status, flight_hr):
        self.name = name
        self.flight_hr = flight_hr
        self.status = status
        
    def __repr__(self):
        return f'{self.name}, {self.status}'
        
    def __gt__(self, other):
        if self.flight_hr > other.flight_hr:
            return True
        return False
    
    def __ge__(self, other):
        if self.flight_hr >= other.flight_hr:
            return True
        return False
    
    def __eq__(self, other):
        if self.flight_hr == other.flight_hr:
            return True
        return False
        
    def name(self):
        return name
    
    def flight_hr(self):
        return flight_hr
    
    def status(self):
        return status
```

### Build a list of astronaut objects

 - List comprehension used for conciseness
 - Used the "next" operator to dump the first value in the list since it is column headers

```python
next(data)
astronaut_objects = [Astronauts(row[0], row[3], row[13]) for row in data]
```

### List all mutable objects for first astronaut object

```python
vars(astronaut_objects[0])
```

### Pick two random astronaut objects and compare them
 - I used a loop to display several different groups of two astronauts

```python
import random

for i in range(3):
    passed = False
    while not passed:
        random_astronaut_1 = random.choices(astronaut_objects)[0]
        random_astronaut_2 = random.choices(astronaut_objects)[0]
        if random_astronaut_1 != random_astronaut_2:
            passed = True
            
    print(random_astronaut_1) # This is displaying the result of using the __repr__ method
    print(f'\nAstronaut 1: {random_astronaut_1.name}\nFlight Hours: {random_astronaut_1.flight_hr}\n')
    print(random_astronaut_2) # This is displaying the result of using the __repr__ method
    print(f'\nAstronaut 2: {random_astronaut_2.name}\nFlight Hours: {random_astronaut_2.flight_hr}\n')
    print('Does Astronaut 1 have more flight hours than astronaut 2?', random_astronaut_1 > random_astronaut_2)
    print("\n--------------------------------------------------------------------------------")
    
```
